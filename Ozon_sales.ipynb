### Создание функций для просмотра информации о данных:

```python
def table_info(table):
    # Вывод информации о таблице
    print('\033[1m' + 'Информация о таблице:' + '\033[0m')
    print(table.info())
    
    # Проверка наличия дубликатов
    print('\033[1m' + 'Количество явных дубликатов в таблице:' + '\033[0m')
    print(table.duplicated().sum())
```

```python
def pass_value_barh(df):
    # Вычисляем процент пропусков для каждого столбца
    missing_data = (df.isna().mean() * 100).to_frame(name='space').query('space > 0').sort_values(by='space')
    
    # Проверяем, есть ли пропущенные значения
    if not missing_data.empty:
        ax = missing_data.plot(kind='barh', figsize=(10, 3), rot=0, legend=False, fontsize=14, color='skyblue')
        ax.set_title('Пропущенные значения, %', fontsize=22, color='SteelBlue')
        ax.set_xlabel('Процент пропусков', fontsize=18)
        ax.set_ylabel('Столбцы с пропусками', fontsize=18)
        ax.set_xlim(0, 100)
        plt.show()
    else:
        print('Пропусков не осталось, или данных для анализа нет.')
```

```python
# Функция для извлечения названия города
def extract_city(warehouse_name):
    if '_РФЦ' in warehouse_name:
        city = warehouse_name.split('_РФЦ')[0]
    elif '_МРФЦ' in warehouse_name:
        city = warehouse_name.split('_МРФЦ')[0]
    else:
        city = warehouse_name
    city = city.lower().capitalize()
    return city
```

```python
def extract_color_v4(product_name):
    # Ищем цвет в строке
    color_match = re.search(r'Цвет:\s*([^,]+)', product_name, re.IGNORECASE)
    if color_match:
        color = color_match.group(1).strip().lower()
        
        # Удаляем цифры и специфические фразы
        color = re.sub(r'\b\d+\b', '', color)  # Убираем числа
        color = re.sub(r'\bбыстросохнущая\b', '', color, flags=re.IGNORECASE)
        color = re.sub(r'\bматовая\b', '', color, flags=re.IGNORECASE)
        color = color.replace('.', '').strip()  # Удаляем точки и лишние пробелы
        
        return color
    return None

df_orders['Цвет'] = df_orders['Наименование товара'].apply(extract_color_v4)
```

```python
# Фильтрация заказов по выбранным регионам
filtered_orders = df_orders[df_orders['Регион доставки'].isin(top_10_region_list)]

# Группировка по регионам и категориям с подсчетом количества заказов
category_preferences = (filtered_orders
                        .groupby(['Регион доставки', 'Первая категория'])['Количество']
                        .sum()
                        .reset_index()
                        .rename(columns={'Количество': 'Количество шт'}))

# Получение уникального списка регионов для построения диаграмм
regions = category_preferences['Регион доставки'].unique()

# Настройка количества строк и столбцов для диаграмм
rows = (len(regions) + 1) // 2  # Рассчитываем необходимое количество строк

# Создание подграфиков
fig, axes = plt.subplots(rows, 2, figsize=(15, 5 * rows))

# Упрощаем доступ к осям
axes = axes.flatten()

for i, region in enumerate(regions):
    region_data = category_preferences[category_preferences['Регион доставки'] == region]
    total_orders = region_data['Количество шт'].sum()
    region_data = region_data.copy()  # Создаем копию, чтобы избежать предупреждений SettingWithCopy
    region_data['Доля'] = (region_data['Количество шт'] / total_orders) * 100
    
    # Фильтрация категорий с долей менее 1%
    region_data = region_data[region_data['Доля'] >= 1]
    
    axes[i].pie(
        region_data['Доля'], labels=region_data['Первая категория'],
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors
    )
    axes[i].set_title(f'{region}')
    axes[i].axis('equal')  # Обеспечивает круговую форму диаграммы

plt.tight_layout()
plt.show
```

```python
# Группировка данных по категориям товаров и периодам, расчет общей суммы продаж и количества проданных товаров
category_sales = df_orders.groupby([df_orders['Принят в обработку'].dt.to_period('M'), 'Первая категория']).agg({
    'Сумма отправления': 'sum',
    'Количество': 'sum'
}).reset_index()

# Расчет средней цены за единицу товара по категориям
category_sales['Средняя цена'] = category_sales['Сумма отправления'] / category_sales['Количество']

# Переименование столбцов для ясности
category_sales = category_sales.rename(
    columns={'Сумма отправления': 'Общая сумма заказа',
             'Принят в обработку': 'Период', 'Количество': 'Количество шт'})
category_sales
```

```python
category_sales['Период'] = category_sales['Период'].astype(str)

# Исключение категории 'Аромадиффузор' из анализа, так как он продавался только в июле
categories_filtered = category_sales['Первая категория'].unique()
categories_filtered = [category for category in categories_filtered if category != 'Аромадиффузор']

# Построение графиков изменения количества продаж и средней цены для каждой категории
fig, axes = plt.subplots(len(categories_filtered), 2, figsize=(14, len(categories_filtered) * 4))

for i, category in enumerate(categories_filtered):
    category_data = category_sales[category_sales['Первая категория'] == category]
    
    # График изменения количества продаж
    sns.lineplot(x='Период', y='Количество шт', data=category_data, ax=axes[i, 0], marker='o')
    axes[i, 0].set_title(f'Изменение количества проданных шт: {category}')
    axes[i, 0].set_xlabel('Период')
    axes[i, 0].set_ylabel('Количество шт')
    
    # График изменения средней цены
    sns.lineplot(x='Период', y='Средняя цена', data=category_data, ax=axes[i, 1], marker='o')
    axes[i, 1].set_title(f'Изменение средней цены: {category}')
    axes[i, 1].set_xlabel('Период')
    axes[i, 1].set_ylabel('Средняя цена')

plt.tight_layout()
plt.show()
```

- **Цвета для продвижения**: **Белоснежный** здесь пользуется большей популярностью, чем **Ванильный мусс**, а **Нежный пепел** имеет значительную долю продаж.

#### Ростовская Область:
- **Товары для продвижения**: В линейке **Velvet** нет явного лидера между фасовками 1,5 кг и 3,8 кг, что позволяет гибко подходить к их продвижению. **MIA 830 гр** также имеет высокий процент продаж, что указывает на интерес к большим фасовкам.
- **Цвета для продвижения**: **Ванильный мусс** остается лидером, за ним следуют **Пудра**, **Черный** и **Белый**. Важно учитывать это разнообразие при планировании рекламных кампаний.

#### Самарская Область:
- **Товары для продвижения**: Высокая доля больших фасовок у красок для мебели, таких как **Наборы** и **MIA 830 гр**. В линейке **Velvet** 1,5 кг пользуются большей популярностью, но **Velvet 7 кг** также имеет высокий процент продаж. Необходимо проверить видимость и доступность товаров, таких как **MIA 830 гр** и **Velvet 7 кг**. **Сопутка** занимает высокую долю, что может быть использовано для продвижения красок для мебели через адгезионный грунт.

#### Челябинская Область:
- **Товары для продвижения**: Высокий процент продаж **MIA 1 л** делает этот товар приоритетом для рекламы. **Velvet 3,8 кг** имеет крайне низкие продажи, что требует проверки наличия и доступности товара.
- **Цвета для продвижения**: **Белый** остается лидером продаж, но также стоит отметить предпочтение к цветам **Тирамису**, **Тауп** и **Маренго**.

#### Приморский Край:
- **Товары для продвижения**: Несмотря на низкую скорость доставки (13 дней), продажи в регионе превышают 300,000 руб. Продвигайте **MIA 400 гр** и **Velvet 3,8 кг**, чтобы увеличить объем продаж.
- **Цвета для продвижения**: Фокусируйтесь на **Белоснежный** и **Ванильный мусс**, как наиболее популярные цвета.
- **Особенности региона**: Важно проанализировать покрытие складов и возможности отгрузки группы А и А+ в этот регион для повышения скорости доставки и увеличения продаж.

#### Общие рекомендации:
1. **Скорость доставки**: Приморский край имеет более 300.000 продаж, хотя у него достаточно низкая скорость доставки - 13 дней. Важно посмотреть какие склады покрывают данный регион и отгрузить группу А и А+ туда, возможно сумма продаж этого региона сильно увеличится, когда они смогут быстрее получать товар. Ямало-Ненецкий Автономный округ, Красноярский Край и Республика Коми тоже имеют высокий уровень продаж, несмотря на среднюю скорость доставки в 7 дней. Стоит рассмотреть покрытие складов и этих регионов. 
2. **Процент отмен**: Большинство регионов с высоким чеком имеют в среднем от 7 до 8% доли отмененных заказов. Ленинградская область имеет самый низкий процент 4,8%, вероятно это связано с нахождение склада в СПб, наличием FBS и высокой скоростью доставки. Важно уделить внимание Челябинской области, у нее самая высокая доля отмен (12,9%) из рассматриваемых регионов с высоким средним чеком. Она имеет не самую низкую среднюю скорость доставки (3 дня). Лучше более глубоко проанализировать причины таких частых отмен в данном регионе. 


### Способы оплаты

Доля продаж по Ozon карте составляет 67%, можно попробовать через внутренние маркетинговые инструменты дать больше скидку владельцам этих карт, чтобы простимулировать продажи.

### Ценовые диапозоны

Так как основная часть продаж сосредоточена в нижнем ценовом сегменте до **2000 рублей**, с пиком около **1000 рублей**, а в диапозонах (от **2000 до 4000 рублей**) спрос заметно снижается, можно рассмотреть представить больше товаров в данном диапозоне и попробовать продвигать их для повышения среднего чека.

### Эластичность спроса

В каких категориях снижение цены повлияло на увеличение продаж? MIA 1л, Velvet 1.5 кг, Velvet 3.8 кг, Velvet 7 кг. Это наиболее заметно у товаров дороже 2000р.

В каких категориях увеличение цены не повлияло на снижение продаж? MIA 400гр, Сопутка MIA. Спрос на данные категории стабилен и даже повышение цен в условиях повышенных продаж в сезон не повлиял на объем проданных шт данных категорий. 

В каких категориях увеличение цены повлияло на снижение продаж? У MIA 830гр наблюдается падение в июне при увеличении цены, однако при условии повышенного трафика и увеличение сезонности товара в июле, еще более сильное повышение цены в июле не повлияло на падение продаж данной категории. Грунт концентрат: имеет прямую зависимость продаж от увеличения или снижения цены. 

В каких категориях снижение цены снизило продажи? Неоднозначная картина по Velvet тестер, так как снижение в стоимости не такое большое, но снижение цены в июне привело к снижению продаж, а в июле наоборот, увеличение продаж при условии снижении цены. Здесь нет прямой зависимости с ценой, так как в июле в целом наблюдается повышенный спрос товаров, но важно в дальнейшей более детально проследить за влиянием цены на продажи у данной категории.

### Сочетаемость товаров

Касательно комплектации заказов нет каких-то невероятных открытий. Покупка товаров разных объемов в одном заказе или сочетание покупки краски и к ней грунта является стандартной картиной в данной нише. Однако есть интересный момент, что с товаром Mia 400 гр достаточно часто покупают Velvet тестер, что может говорить о интересе клиентов, которые покупают краски для мебели, попробовать краску для стен через тестер. Если в ассортименте товара имелись бы тестеры красок для мебели, возможно мы могли бы наблюдать картинку покупки тестера Mia с линейкой для стен Velvet. Можно попробовать более явно показать клиентам возможность покупки тестера в карточках товара красок для мебели с помощью инструментов визуализации на инфографике или других маркетинговых инструментов Озон (например: добавить в подборку “Серия от бренда”). Также можно давать тестер в подарок при сборке больших заказов с краской для стен по системе ФБС ( т.к. по ФБО это невозможно контролировать). Или добавить тестеры как подарок в товары красок для мебели в коробках с высокой стоимостью (Наборы по 2 или 3 шт).
